```python
def extractImages(pathIn):
    area = (420,0,1500, 1080)
    for video in pathIn: 
        count=0
        vidcap = cv2.VideoCapture(video)
        success,image = vidcap.read()
        success = True
        while success:
            vidcap.set(cv2.CAP_PROP_POS_MSEC,(count*300))    # added this line 
            success,image = vidcap.read()
            if(not success):
                break
            name= ( "ddubi%d.jpg" % count)
            print(name)
            cv2.imwrite('./capture/'+'ddubi%d.jpg' % count, image)     # save frame as JPEG file
            img = cv2.imread('./capture/'+name)
            img = cv2.resize(img, (500, 300)) 
            img = imutils.rotate_bound(img, 90)
            cv2.imwrite('./images/'+name, img)
            count = count + 1

# if __name__=="__main__":
#     print("aba")
#     a = argparse.ArgumentParser()
#     a.add_argument("--pathIn", help="path to video")
#     a.add_argument("--pathOut", help="path to images")
#     args = a.parse_args()
#     print(args)
#     extractImages(args.pathIn, args.pathOut)
```

```python
# grab the dimensions of the image and calculate the center
# of the image
image = cv2.imread("2218_1.jpg")
(h, w) = image.shape[:2]
center = (w / 2, h / 2)
 
# rotate the image by 180 degrees
M = cv2.getRotationMatrix2D(center, 270, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))
cv2.imwrite("rotated.jpg", rotated)
```
